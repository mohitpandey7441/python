# Match Case Statement

```python
number = 2

match number:
    case 1:
        print("One")
    case 2 | 3:
        print("Two or Three")
    case _:
        print("Other number")
```

```python
day = 4

match day:
    case 1:
        print('Monday')
    case 2:
         print('Tuesday')
    case 3:
        print('Wednesday')
    case 4:
        print('Thurseday')
    case 5:
        print('Friday')
    case 6:
        print('Saturady')
    case 7:
        print('Sunday')
        
    case _:
         print('Out of day')

          
```

# Using input function

```python

day = int(input('Enter a day number'))

match day:
    case 1:
        print('Monday')
    case 2:
         print('Tuesday')
    case 3:
        print('Wednesday')
    case 4:
        print('Thurseday')
    case 5:
        print('Friday')
    case 6:
        print('Saturady')
    case 7:
        print('Sunday')
        
    case _:
         print('Out of day')
```
