# SuperSet , SubSet , Disjoint

```python
a = {1, 2, 3, 4, 5, 6, 7, 8, 9}
b = {3, 4, 5, 6, 7, 8}
c = {10 , 20, 30, 40}
```

```python
a.issuperset(b)
```

```python
b.issuperset(a)
```

```python
b.issubset(a)
```

```python
c.issubset(a)
```

```python
print(a)
print(b)
print(c)
```

```python
c.isdisjoint(a)
```

```python
c.isdisjoint(b)
```
