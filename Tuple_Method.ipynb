# Tuple


```python
# Tuple is a sequence of value is similar to list.
# Features: 1: ordered , 2: immutable( not change)
# 3. duplicate value are allowed, 4: can store different data type
```

# Tuple creation

```python
tup1 = () # Empty tuple
tup1
```

```python
type(tup1)
```

```python
tup2 = (1, 23,45) # Integer number
tup2
```

```python
tup3 = (12.3 , 3.4, 5.6) # Float number
tup3
```

```python
tup4 = ("one","two", "three") #  string tuple
tup4
```

```python
tup5 = ("Asif" , 25 , (50, 100),(150 , 90)) # Nested tuple
tup5
```

```python
tup6 = (12, 3.4 , 'abcd' , True) # Mixed data type
tup6
```

```python
len(tup7) # Length of list
```

# Tuple Slicing

```python
mytuple = ('one', 'two', 'three', 'four','five', 'six', 'seven', 'eight')
```

```python
mytuple[0:3] # Return all item of 0th index to 3rd index location excluding the item
```

```python
mytuple[2:5] # List all item of 2nd index to 5th index location excluding the item
```

```python
mytuple[:3] # Return first three element
```

```python
mytuple[:2] # Return first two element
```

```python
mytuple[-3:] # Return last three element
```

```python
mytuple[-1:] # Return last item of tuple
```

```python
mytuple[::2] 
```

```python
mytuple[:] # Retuen whole tuple
```

```python
mytuple[::] # Return whole tuple
```
