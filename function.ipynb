
# Function

```python
# Define a function

def fun(): # Without arguments
    print('welcome to gfg')
```

```python
def fun():
    print('Good Morning')

fun()  # calling the function  
```

```python
def fun(x,y):
    c = x + y
    print(c)

fun(5,6)    
```

```python
def evenOdd(x):
    if(x % 2==0):
        return('Even')
    else:
        return('Odd')

print(evenOdd(16))
print(evenOdd(7))
        
```

# Default Arguments

```python
def myfun(x, y=50):
    print("x:",x)
    print("y:",y)

myfun(10)    
```

```python
def student(fname, lname):
    print(fname, lname)

student(fname='Geeks', lname='Practice')
student(lname='Practice', fname='Geeks')
```
